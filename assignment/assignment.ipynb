```python
#Q2.1
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv('/content/airbnb_hw.csv', low_memory=False)
print( df.shape, '\n')
df.head()
```

```python
price = df['Price']
price.unique() #lists as an array with commas in between - we want it cleaner
```

```python
price = df['Price']
price = price.str.replace(',','')
print( price.unique() , '\n')
price = pd.to_numeric(price,errors='coerce')
print( price.unique() , '\n')
print( 'Total missing: ', sum( price.isnull() ) )
# now we replaced the commas with nothing and got rid of the apostrophe so it looks cleaner
# we end up with no missing values
```

```python
#Q2.2
df = pd.read_csv('/content/sharks.csv', low_memory=False)
df['Type'].value_counts()  # get the value count for each type
```

```python
type = df['Type'] #temporary vector
type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # gets all watercraft/boating values
type.value_counts()
type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # replaces all unclean values
type.value_counts()
df['Type'] = type # replace type variable with the cleaned version
del type # gets rid of temporary vector
df['Type'].value_counts() # gives the value counts for unprovoked shark attacks, provoked shark attacks, and watercraft accidents
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan)
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y')
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index') #gives probability of fatal accidents (yes or no)
```

```python
#Q2.3
df = pd.read_parquet('/content/justice_data.parquet')
release = df['WhetherDefendantWasReleasedPretrial']
print(release.unique(),'\n')
print(release.value_counts(),'\n')
release = release.replace(9,np.nan)
print(release.value_counts(),'\n')
sum(release.isnull())
df['WhetherDefendantWasReleasedPretrial'] = release
del release
```

```python
#Q2.4
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
length = pd.to_numeric(length,errors='coerce')
length_NA = length.isnull()
print( np.sum(length_NA),'\n') # we have missing values
print( pd.crosstab(length_NA, type), '\n')
length = length.mask( type == 4, 0) # replacing length
length = length.mask( type == 9, np.nan) # replacing length
length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # less are missing
df['ImposedSentenceAllChargeInContactEvent'] = length # replacing data with cleaned version
del length, type
```
